```python
from bs4 import BeautifulSoup
import numpy as np
import requests 
from urllib.request import urlopen
```

```python
#url = "https://movie.naver.com/movie/bi/mi/point.nhn?code="+
html=list(range(10))
for n in range (10) :
    url = 'https://movie.naver.com/movie/bi/mi/point.nhn?code='+str(167600+n)
    #url = base_url.format(n+1)
    #webpage = urlopen(url)
    resp = requests.get(url)
    html[n] = BeautifulSoup(resp.content, 'html.parser')
html[5]
```

```python
genre_list = ["드라마", "판타지", "서부", "공포", "멜로/로맨스", "모험",
              "스릴러", "느와르", "컬트", "다큐멘터리", "코미디", "가족",
              "미스터리", "전쟁", "애니메이션", "범죄", "뮤지컬", "SF", "액션",
              "무협", "에로", "서스펜스", "서사", "블랙코미디", "실험", "공연실황"]
```

```python
grp_point = html[5].find('ul', {'class': 'grp_point'})
if grp_point == None :
    print(0)
else:
    grp_list = grp_point.findAll('span')
genre = html[5].find('dl', {'class': 'info_spec'}).findAll(string=genre_list) 
#장르 리스트에 있는 것들이 포함된 문자열을 찾아서 출력
```

```python
for i in range (5):
    grp_list[i]=grp_list[i].string #숫자만 출력
grp_list
```

```python
netizen_score = html.find('div', {'id': 'netizen_point_tab_inner'})
scr_list=netizen_score.findAll('em')
k=0.1 # 계산을 위한 값
scr_list[0]=float(scr_list[0].string) #숫자로 변환
score = scr_list[0]
for i in range (2):
    scr_list[i+1]=scr_list[i+2] #기존에 있는 값을 뒤에있는 값으로 대체
    scr_list[i+1]=float(scr_list[i+1].string) #숫자만 출력
    score = score + scr_list[i+1] * k
    k=k/10
score
```
